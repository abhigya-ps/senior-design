```python
import csv
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
```

```python
# RA_RA12S-701C37_2022-08-04-1406-1day

with open('RA_RA12S-701C37_2022-08-04-1406-1day.csv', 'r') as file:
    reader = csv.reader(file)
    rows = list(reader)
```

```python
temperatures = [row[4] for row in rows][1:]
temperatures = [float(temp) for temp in temperatures]
timestamps = [float(i) for i in range(len(temperatures))]

plt.plot(temperatures)
```

```python
# test 1 - first five temperatures
x = np.array(timestamps[0:5])
y = np.array(temperatures[0:5])

# scatterplot without regression line
plt.plot(x, y, 'o')
```

```python
# scatterplot with regression line
plt.plot(x, y, 'o')
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b)
```
