```python
import csv
import matplotlib.pyplot as plt
from collections import deque
import numpy as np
import time
from termcolor import colored
import pylab as pl
from IPython import display
```

```python
# load data from the previous week
with open('RA_RA12S-701C37_2022-08-02-5days.csv', 'r') as file:
    reader = csv.reader(file)
    rows = list(reader)

timestamps = [row[0] for row in rows][1:]

temperatures = [row[4] for row in rows][1:]
temperatures = [float(temp) for temp in temperatures]  # converting temperatures in string to float
```

```python
# calculate metrics for temperatures from the last 5 days

metrics = {}
metrics["median"] = np.percentile(temperatures, 50)
metrics["sd"] = np.std(temperatures)
metrics["lower_bound"] = metrics["median"] - 2 * metrics["sd"]
metrics["upper_bound"] = metrics["median"] + 2 * metrics["sd"]
metrics["ninety_pct"] = np.percentile(temperatures, 90)
metrics["ten_pct"] = np.percentile(temperatures, 10)
metrics["seventy_five_pct"] = np.percentile(temperatures, 75)
metrics["twenty_five_pct"] = np.percentile(temperatures, 25)

# rounding of the metrics to 3 decimal places
for key, value in metrics.items():
    metrics[key] = round(value, 3)
    
print("Metrics for temperatures from the last 5 days: \n", metrics)
plt.plot(temperatures)
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Temperatures from 07/28 to 08/02')
```

```python
# function that determines alert level
def classify_temp(temp, metrics):
    # unpacking metrics
    median = metrics["median"]
    sd = metrics["sd"]
    lower_bound = metrics["lower_bound"]
    upper_bound = metrics["upper_bound"]
    ninety_pct = metrics["ninety_pct"]
    ten_pct = metrics["ten_pct"]
    seventy_five_pct = metrics["seventy_five_pct"]
    twenty_five_pct = metrics["twenty_five_pct"]
    
    if temp < lower_bound or temp > upper_bound:
        return "Out of Bounds"
    elif temp < ten_pct or temp > ninety_pct:
        return "Red"
    elif temp < twenty_five_pct or temp > seventy_five_pct:
        return "Yellow"
    else:
        return "Green"
```

```python
# load data for today
with open('RA_RA12S-701C37_2022-08-02-1day.csv', 'r') as file:
    reader = csv.reader(file)
    rows = list(reader)

recent_temps = [row[4] for row in rows][1:]
recent_temps = [float(temp) for temp in recent_temps]  # converting temperatures in string to float
print(type(recent_temps[0]))

plt.plot(recent_temps)
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Temperatures for today 08/02')
```

```python
# make a classification on the new temperature

# 'lower_bound': 78.271, 'upper_bound': 85.449
color_map = {"Red": "red", "Yellow": "yellow", "Green": "green", "Out of Bounds": "magenta"}

# test 1 - green alert when temperature is between 25th and 75th percentiles
temp1 = recent_temps[0]
alert_level = classify_temp(temp1, metrics)
color = color_map[alert_level]
print(colored(f"Alert for {temp1} F - {color.upper()}", color))

# test 2 - yellow alert when temperature is within 10th and 90th percentiles but not outside the range for green alert
temp2 = recent_temps[13]
alert_level = classify_temp(temp2, metrics)
color = color_map[alert_level]
print(colored(f"Alert for {temp2} F - {color.upper()}", color))

# test 3 - red alert when temperature is beyond 10th and 90th percentiles
temp3 = recent_temps[15]
alert_level = classify_temp(temp3, metrics)
color = color_map[alert_level]
print(colored(f"Alert for {temp3} F - {color.upper()}", color))
```

```python
# determining alert levels for all temperatures from today
%matplotlib inline

color_map = {"Red": "red", "Yellow": "yellow", "Green": "green", "Out of Bounds": "magenta"}
queue = deque()   # double-ended queue that stores the last 5 differences in temperature
total = 0
n = len(recent_temps)
for i in range(n):
    # plot new temperature value
    temp = recent_temps[i]
    alert_level = classify_temp(temp, metrics)
    color = color_map[alert_level]
    pl.clf()
    pl.plot(recent_temps[:i+1], color, linewidth=2)
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.title('Temperatures for today 08/02')
    display.clear_output(wait=True)
    display.display(pl.gcf())
    print(f"Alert level for {temp} degrees: {alert_level}!")
    time.sleep(1)
```
