```python
import csv
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('RA_RA12S-701C51_2022-07-26-1440.csv', 'r') as file:
    reader = csv.reader(file)
    rows = list(reader)

print(rows[0])
print(rows[0][0], rows[0][4])
```

```python
timestamps = [row[0] for row in rows][1:]
print(len(timestamps))
```

```python
temperatures = [row[4] for row in rows][1:]
n = len(temperatures)
timestamps = [i for i in range(n)]
print(temperatures)
plt.plot(timestamps, temperatures)
```

```python
# calculate metrics for data from the last 5 days

# convert temperatures from string to float
temperatures = [float(temp) for temp in temperatures]

# find the min and max
min_temp = float('inf')
max_temp = float('-inf')
for temp in temperatures:
    min_temp = min(min_temp, temp)
    max_temp = max(max_temp, temp)

print("Lowest temperature in the last 5 days:", min_temp)
print("Highest temperature in the last 5 days:", max_temp)
```

```python
# calculating the median

median = np.percentile(temperatures, 50)
sd = np.std(temperatures)
lower_bound = median - 2 * sd
upper_bound = median + 2 * sd
ninety_pct = np.percentile(temperatures, 90)
ten_pct = np.percentile(temperatures, 10)
seventy_five_pct = np.percentile(temperatures, 75)
twenty_five_pct = np.percentile(temperatures, 25)
print(median)
print(sd)
print(lower_bound)
print(upper_bound)
print(ninety_pct)
print(ten_pct)
print(seventy_five_pct)
print(twenty_five_pct)
```
